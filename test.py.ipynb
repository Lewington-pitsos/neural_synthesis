```python
from utils import pyramid, img
```

```python
img_path = "data/samples/beksinski-layers5-blur6-scale_decay0-scale100000-optimizerLBFGS-lr0.1/"
img_name = "256.jpg" # we use this variable later
full_path = img_path + img_name

example_texture = img.image_loader(full_path)
img.show_image(example_texture)
```

```python
gauss_kernel = pyramid.build_gauss_kernel(size=3, sigma=0.01, n_channels=3, cuda=True)
blurred = pyramid.conv_gauss(example_texture, gauss_kernel)
img.show_image(blurred)
```

```python
pyr = pyramid.gaussian_pyramid(example_texture, gauss_kernel, max_levels=4)
for i in pyr:
    img.show_image(i)
```
